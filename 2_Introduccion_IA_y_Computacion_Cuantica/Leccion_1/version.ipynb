```python
from sklearn.linear_model import LinearRegression
import numpy as np

X = np.array([1,2,3,4,5]).reshape(-1, 1)
Y = np.array([2,4,6,8,10])

# Crear el modelo de regresión lineal
model = LinearRegression()
# Entrenar el modelo
model.fit(X, Y)

# Hacer una predicción
print(model.predict([[6]]))
```

```python
#Circuito Cuántico Simple en Qiskit
from qiskit import QuantumCircuit
from qiskit_aer import Aer

# Crear un circuito cuántico
qc = QuantumCircuit(1,1) # primer 1 es el número de qubits, el segundo 1 es el número de bits clásicos
qc.h(0)  # Aplicar una puerta Hadamard
qc.measure(0,0)
#simular usando Aer simulator
simulator = Aer.get_backend('aer_simulator')
#Correr el circuito
job = simulator.run(qc)
#Obtener los resultados
result = job.result()

#imprimir los resultados
print(result.get_counts())
```
