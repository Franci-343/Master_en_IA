```python
import pandas as pd
#pandas ayuda a manipular y analizar datos de manera eficiente

# Crear un DataFrame
datos = {'nombre': ['Ana', 'Luis', 'Carlos', 'Marta'],
         'edad': [28, 34, 29, 42],
         'salario': [3500, 4200, 3900, 5000]}
df = pd.DataFrame(datos)
print("DataFrame:\n", df)
```

```python
df.fillna(0, inplace=True) #rellena los valores nulos con 0
filtered_df = df[df['edad'] > 30] #filtra las filas donde la edad es mayor a 30
print("Filtrado por edad > 30:\n", filtered_df)
```
