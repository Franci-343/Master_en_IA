```python
#Pre procesamiento de datos en pandas
import pandas as pd
from sklearn.preprocessing import StandardScaler, OneHotEncoder

#Conjunto de datos de ejemplo
datos = {
    'edad': [25, 32, None, 51, 62],
    'salario': [50000, 60000, None, 120000, 150000],
    'genero': ['Masculino', 'Femenino', 'Femenino', 'Masculino', 'Femenino']
}
df = pd.DataFrame(datos)

df["edad"] = df["edad"].fillna(df["edad"].mean())
df["salario"] = df["salario"].fillna(df["salario"].mean())

encoder = OneHotEncoder(sparse_output=False)
genero_encoded = encoder.fit_transform(df[['genero']])
genero_df = pd.DataFrame(genero_encoded, columns=['genero_Femenino', 'genero_Masculino'])

scaler = StandardScaler()
df[['edad', 'salario']] = scaler.fit_transform(df[['edad', 'salario']])
df_final = pd.concat([df[['edad', 'salario']], genero_df], axis=1)
print("DataFrame preprocesado:\n", df_final)
```
