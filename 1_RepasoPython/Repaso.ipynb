```python
x = 10
y = 20
if x < y:
    print(f"{x} es menor que {y}")
else:
    print(f"{x} es mayor o igual que {y}")

a = 5
b = 15
if a > b:
    print(f"{a} es mayor que {b}")
elif a < b:
    print(f"{a} es menor que {b}")
else:
    print(f"{a} es igual a {b}")
```

##Bucles for, while y do while

```python
print("Bucle for:")
for i in range(5):
    print(i)
print("Bucle while:")
contador = 0
while contador < 5:
    print(contador)
    contador += 1
print("Bucle do while:")
contador = 0
while True:
    print(contador)
    contador += 1
    if contador >= 5:
        break
```
