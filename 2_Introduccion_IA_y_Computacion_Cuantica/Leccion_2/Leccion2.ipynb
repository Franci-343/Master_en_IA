```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

X = np.array([500,700,900,1100,1300]).reshape(-1, 1)
Y = np.array([1500,2100,2900,3500,4300])

model = LinearRegression()
model.fit(X, Y)

predicted_price = model.predict([[1200]])
print(f"Predicted price for a house with 1200 sq ft: ${predicted_price[0]:.2f}")

plt.scatter(X, Y,color='green',label='Actual Prices')
plt.plot(X, model.predict(X), color='blue', label='Regression Line')
plt.xlabel("House Size (sq ft)")
plt.ylabel("Price ($)")
plt.title("Linear Regression House Price prediction")
plt.legend()
plt.show()
```
