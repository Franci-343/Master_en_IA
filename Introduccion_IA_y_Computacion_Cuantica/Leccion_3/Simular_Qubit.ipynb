```python
#Simularemos un qubit usando python y qiskit
from qiskit import QuantumCircuit
from qiskit_aer import Aer
from qiskit.visualization import plot_histogram
import matplotlib.pyplot as plt

# Crear un circuito cu치ntico con 1 qubit y 1 bit cl치sico
qc = QuantumCircuit(1, 1)

qc.h(0) # Aplicar una puerta Hadamard para poner el qubit en superposici칩n

qc.measure(0, 0) # Medir el qubit y almacenar el resultado en el bit cl치sico

# Simular el circuito usando el simulador Aer
simulator = Aer.get_backend('aer_simulator')
#correr el circuito
job = simulator.run(qc, shots=1000) # Ejecutar el circuito 1000 veces
# Obtener los resultados
result = job.result()
counts = result.get_counts(qc)
print(counts)

labels = list(counts.keys())
values = list(counts.values())

# Graficar los resultados
colors = ['green', 'red']

plt.bar(labels, values, color=colors)
plt.xlabel('Measurement Outcome')
plt.ylabel('Count')
plt.title('Quantum Measurement Results')
plt.show()
```
